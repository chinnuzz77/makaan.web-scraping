```python
import time 
import requests 
from IPython.display import clear_output 
base_url = 'https://www.makaan.com/hyderabad-residential-property/buy-property-in-hyderabad-city?page=' 
ow = []   
bhk = [] 
p = [] 
s = [] 
loc = [] 
st = [] 
pps = [] 
for i in range(1,51):   
  time.sleep(3)   
  print(f'Going to scrape data from {i} page')  
  clear_output(wait=True)   
  url = base_url+str(i)+'&_=1624455000063'     
  req = requests.get(url)   
  soup = BeautifulSoup(req.text,'html')  
    #owner info  
  ow_info = soup.findAll('span',attrs={'class':'seller-type'})   
  for j in ow_info:    
   ow.append(j.text)     
        #bhk info  
  bhk_info = soup.findAll('div',attrs={'class':'title-line'})  
  for j in bhk_info:    
   bhk.append(j.strong.span.text) 
               #price info  
  p_info = soup.findAll('div',attrs={'data-type':'price-link'})  
  for  j in p_info:     
   p.append(j.text)     
                    #size info   
  s_info = soup.findAll("td",attrs={'class':'size'})   
  for j in s_info:
   s.append(j.text)
                   #location info
  loc_info = soup.findAll('span',attrs={'itemprop':'addressLocality'})
  for j in loc_info:
   loc.append(j.text)
                     #status info
  st_info = soup.findAll('tr',attrs={'class':'hcol w44'})
  for j in st_info:
   st.append(j.text)
                      #price per sqft info
  pps_info = soup.findAll('td',attrs={'class':'lbl rate'})
  for j in pps_info:
   pps.append(j.text)

                      
                       
```

```python
data = pd.DataFrame({'Owner_info' : ow,
                     'Bhk_info' : bhk,
                     'price' : p,
                     'Locality' : loc,
                     'Area_sqft' : s,
                     'status' : st,
                     'price_per_sqft' : pps})
data
   
```

```python
data.to_csv('appartmentdata.csv',index=False)
```
